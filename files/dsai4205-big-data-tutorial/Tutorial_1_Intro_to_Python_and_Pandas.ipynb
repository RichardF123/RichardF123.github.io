# Python

Python is a popular programming language in the field of Data Science and AI. It is a high level programming language and has an extensive collection of libraries that can be easily installed. **We require that you use python for this course, so it is important to be familiar with it.**

Python does not utilize curly braces like C, C++, or Java. Instead, it relies on whitespace (tabs or spaces) to determine the scope of methods, loops, and other structures. Typically, we use four spaces to represent one level of indentation. It's important to be consistent with the number of spaces; mixing spaces and tabs will result in a syntax error.

````python
for i in range(10):
    print(i * 2)      # Indent by 4 spaces
    print(i * 4)
    for j in range(4):
        print(i * j)  # Indent 8 spaces
print('Done')       # No indent, runs after for loop
````

## If statement

````python
# Python doesn't require parentheses
a, b = 5, 10

# equality
if a == 5:
    print('yes')
else:
    print('no')

# not equal
if a != 8:
    print('neq')

# and
if a > 5 and b <= 10:
    print('1')
elif a > 10:
    print('2')
else:
    print('3')

# or
if a < 10 or b > 100:
    print('yes')

# Strings also use '=='
s = 'abc'
if s == 'abc':
    print('here')

t = 'abab'
if s in t:
    print("bc")
````

## String manipulation

````python
dept = "Data Science and AI"

print(len(dept)) #print out the length of the string, the number of characters

#standard indexing
print(dept[0]) #print out the 1st character of the string
print(dept[1]) #print out the 2nd character of the string
print(dept[10]) #print out the 11th character of the string

#Slicing "Continuous segment of Data"
print(dept[0:4]) #print out a slicing from the beginning up to the 4th character
print(dept[5:]) #print out a slicing from the 6th character to the end.
print(dept[:4]) #print out a slicing from the beginning up to the 4th character
print(dept[5:100]) #print out a slicing from the 6th character to the end.

#Negative indexing
print(dept[-1]) #print out the last character
print(dept[-2]) #print out the second last character
print(dept[:-2]) #print out a slicing from the beginning up to, but not including, the last two elements.
print(dept[:-4]) #print out a slicing from the beginning up to, but not including, the last four elements.
print(dept[::-2]) #print out a slicing that take every second element in the reverse order.

# Concatentation
print(dept + ' PolyU')
# print(dept + 123) #Error, type mismatch
print(dept + str(123))

# Splitting
print(dept.split('and')) #['Data Science ', ' AI']
print(" AI    ".strip()) #print out the string that remove all leading and trailing spaces
````

## For Loop

````python
a = [1, 2, 3, 4, 5]

range(5)     # Dynamically generates 0, 1, 2, 3, 4
range(1, 5)  # 1, 2, 3, 4

# "Standard" for loop
for i in range(len(a)):
    print(a[i])

# Iterate over each element in the list
for x in a:
    print(x)

# Iterating through dictionaries
d = {0: 'a', 1: 'b', 2: 'c'}
for key, value in d.items():
    print(str(key) + ' -> ' + value)

# Loop over items in an array and get the index at the same time
for index, item in enumerate(a):
    print(index, item)
````

## List Comprehension

A concise way to create lists in Python. It allows you to generate a new list by applying an expression to each item in an iterable, such as a list or range, optionally filtering items with a condition.



The general syntax is:

```
[ <expression> for <name> in <list> if <filter> ]
```

List comprehension also applies to dictionaries and sets, as follows.

```
{ <expression> for <key>, <value> in <dict> if <filter> }
```

````python
def add_two(value):
    return value + 2

# Standard way to apply a function to each element in the list.
x = [1,2,3,4,5]
y = []
for i in range(len(x)):
    y.append(add_two(x[i]))
print(y)

# This is equivalent to above
y = [add_two(x_i) for x_i in x]
print(y)

# With conditions
y = [add_two(x_i) for x_i in x if x_i > 1]
print(y)

# Dictionary comprehensions can be used to create dictionary easily.
d = {x_i: x_i * 2 for x_i in x} # element key : function to element value for element in data structure
print(d)
````

## Merge Operations

UI25E.jpg

**Basic Syntax:**

`pd.merge(left_dataframe, right_dataframe, left_on="some_column", right_on="some_column", how="left|right|inner|outer")`

Assume we have the following mapping for course instructors

````python
course_instructor_mapping = pd.DataFrame({
    "Course Codes": ["DSAI4205", "DSAI5201", "DSAI4202", "DSAI5902"],
    "Instructors": ["FONG Ken", "YANG Xingyi", "NG Vincent", "KUMAR Ajay"]
})

course_instructor_mapping
````

In a Left Merge we are mostly concerned with data on the LEFT side but we would like to add data from the RIGHT side if it has some of the same courses in this case.

````python
pd.merge(left=course_instructor_mapping, right=programme_offer, left_on="Course Codes", right_on="course_codes", how="left")
````

In a Right Merge we are mostly concerned with data on the RIGHT side but we would like to add data from the LEFT side if it has some of the same courses in this case.

````python
pd.merge(left=course_instructor_mapping, right=programme_offer, left_on="Course Codes", right_on="course_codes", how="right")
````

With an Inner Merge, we chop up both dataframes and only glue the stuff that matches. If a course isn't in both
dataframes, we don't keep it and we don't add NaN's. If no type of join is mentioned, then inner join is the
default join.

````python
pd.merge(left=course_instructor_mapping, right=programme_offer, left_on="Course Codes", right_on="course_codes")
````

With an Outer Merge, we chop up both dataframes and keep everything from both sides. Then we toss in NaN's to fill any blanks.

````python
pd.merge(left=course_instructor_mapping, right=programme_offer, left_on="Course Codes", right_on="course_codes", how="outer")
````
